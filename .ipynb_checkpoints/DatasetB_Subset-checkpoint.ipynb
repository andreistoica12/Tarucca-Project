# Get signals function

```python
def get_signals(dataset_path = None, date_list = None):
  d = {0: [], 1: []}

  for date in date_list:
    f = h5py.File( os.path.join(dataset_path, date), "r" )

    if len(f.keys()) == 0:
      continue

    for hour in f.keys():
      runs = f[hour]

      if len(runs.keys()) == 0:
        continue
      
      for run in runs.keys():
        sensors = runs[run][:, [7,8]].T

        #Here we get a subset of the measurements
        #Because it only involves sensors 8,9 there are some assumptions made
        for i, s in enumerate(sensors):
          hit = np.where(s == np.amax(s))[0][0]
          if hit + 70000 < len(s):
            start = hit + 20000
            end = start + 50000
            d[i].append(s[start:end])
          elif hit - 70000 > 0:
            start = hit - 70000
            end = start + 50000
            d[i].append(s[start:end])
          else:
            continue

        
  return d
```

# Automate Subset Extraction

I was unable to run this code because of Google Drive quota being exceeded. I'm assuming that it is limited by Allan's account as the data is sitting on his drive.

```python
STATES = ["undamaged", "15cm_damage", "30cm_damage", "45cm_damage", "repaired"]
SPEEDS = ["40rpm", "lt40rpm", "idle"]
```

```python
def get_datasetB_subset(root_path):

  datasetb_path = os.path.join(root_path, "Data", "Dataset B", "datasetB")
  output_path = os.path.join(root_path, "Data", "Dataset B Subset")

  for state in STATES[:1]:

    for speed in SPEEDS[:1]:

      file_name = "SubsetB_{}_{}_raw_2.csv".format(state, speed)
      datasetb_path_ready = os.path.join(datasetb_path, state, speed)
      output_path_ready = os.path.join(output_path, speed, file_name)
      date_list = os.listdir(datasetb_path_ready)

      print("Get data for {} blade running at {}".format(state, speed))
      data = get_signals(datasetb_path_ready, date_list)

      df = pd.DataFrame.from_dict(data)
      d = {}
      d[0] = np.concatenate([i for i in df[0]])
      d[1] = np.concatenate([i for i in df[1]])
      df = pd.DataFrame.from_dict(d)

      print("Saving data to {}".format(file_name))
```

```python
#@title path to folder: ***AI for Wind Energy share*** {"run": "auto"}
root_path = "/content/drive/MyDrive/AI for Wind Energy share" #@param {"type":"string"}

get_datasetB_subset(root_path)
```
